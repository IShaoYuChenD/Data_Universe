```python
# Cell 1 - imports & utility imports
import sys, os
# ensure project root is on sys.path (we run the notebook from notebooks/ or repo root)
repo_root = os.path.abspath(os.path.join(".."))
if repo_root not in sys.path:
    sys.path.insert(0, repo_root)

from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# import hazard functions (relative import like you used before)
from src.hazard_earthquake import (
    generate_earthquake_events,
    compute_intensity_for_catalog,
    simulate_catalog_and_intensities,
    pga_from_mag_dist,
    haversine_km,
    save_df,
    load_exposures
)
```

```python
# Cell 2 - load exposures from Goal 1

exp_path = "../data/exposure_sample.csv"  
exposures = load_exposures(exp_path)

display(exposures.head())
```

```python
# Cell 3 - determine study bbox from exposures
bbox = {
    'lat_min': float(exposures['latitude'].min()),
    'lat_max': float(exposures['latitude'].max()),
    'lon_min': float(exposures['longitude'].min()),
    'lon_max': float(exposures['longitude'].max())
}
print("Study bbox:", bbox)
```

```python
# Cell 4 - generate an earthquake catalog
n_events = 80
seed = 42
catalog = generate_earthquake_events(n_events=n_events,
                                     region_bounds=bbox,
                                     magnitude_range=(4.5, 7.5),
                                     b=1.0,
                                     depth_km=10.0,
                                     seed=seed)
display(catalog.head())
print("Total events:", len(catalog))
```

```python
# Cell 5 (modified) - compute intensities (catalog -> exposures)
# exposures has columns: id, latitude, longitude, ...
intensities = compute_intensity_for_catalog(
    catalog,
    exposures,
    exposure_id_col='id',
    exposure_lat_col='latitude',
    exposure_lon_col='longitude'
)
display(intensities.head())
print("Intensity rows:", len(intensities))

```

```python
# Cell 6 - aggregate max PGA per exposure and merge back exposure info
# intensities has columns: event_id, exposure_id, distance_km, M, pga_g
# exposures has columns including: id, latitude, longitude, replacement_value, etc.

# first quick sanity
print("Intensity DF shape:", intensities.shape)
print("Exposure DF shape:", exposures.shape)

# compute max PGA per exposure
max_pga = intensities.groupby('exposure_id', as_index=False)['pga_g'].max()
max_pga = max_pga.rename(columns={'pga_g': 'max_pga_g'})

# merge with exposures â€” handle exposure id column name differences robustly
exposure_id_col_candidates = ['id', 'exposure_id', 'exposureId', 'ID']
exp_id_col = next((c for c in exposure_id_col_candidates if c in exposures.columns), None)
if exp_id_col is None:
    raise ValueError("Could not find exposure id column in exposures DataFrame. "
                     "Expected one of: " + ", ".join(exposure_id_col_candidates))

# rename to merge key
exposures_for_merge = exposures.rename(columns={exp_id_col: 'exposure_id'})
merged = max_pga.merge(exposures_for_merge, on='exposure_id', how='left')

print("Merged shape:", merged.shape)
display(merged.head())

```

```python
# Cell 7 - quick stats and inspect top exposures
print("Max PGA summary (g):")
display(merged['max_pga_g'].describe().to_frame().T)

# show top 10 exposures by max PGA
top10 = merged.sort_values('max_pga_g', ascending=False).head(10)
print("Top 10 exposures by max PGA (sample):")
display(top10)
```

```python
# Cell 8 - plot map (scatter); supports columns 'lat' or 'latitude' etc.
# determine latitude/longitude column names in merged
lat_candidates = ['lat', 'latitude', 'y']
lon_candidates = ['lon', 'longitude', 'lng', 'x']

lat_col = next((c for c in lat_candidates if c in merged.columns), None)
lon_col = next((c for c in lon_candidates if c in merged.columns), None)
if lat_col is None or lon_col is None:
    raise ValueError("Could not find lat/lon columns in merged exposures. "
                     "Expected one of: lat/latitude and lon/longitude.")

plt.figure(figsize=(9,6))
sc = plt.scatter(merged[lon_col], merged[lat_col], s=20, c=merged['max_pga_g'])
plt.colorbar(sc, label='Max PGA (g)')
plt.xlabel(lon_col)
plt.ylabel(lat_col)
plt.title('Max PGA per exposure (sample)')
plt.grid(True)
plt.show()
```

```python
# Cell 9 - save outputs for later modules (vulnerability -> losses)
out_root = "../outputs/goal2"
os.makedirs(out_root, exist_ok=True)

catalog_path = os.path.join(out_root, "catalog_goal2.csv")
intensities_path = os.path.join(out_root, "intensities_goal2.csv")
merged_path = os.path.join(out_root, "exposure_max_pga_goal2.csv")

# use the save_df helper if available; fallback to pandas.to_csv
try:
    save_df(catalog, catalog_path)
    save_df(intensities, intensities_path)
    save_df(merged, merged_path)
except NameError:
    catalog.to_csv(catalog_path, index=False)
    intensities.to_csv(intensities_path, index=False)
    merged.to_csv(merged_path, index=False)

print("Saved files:")
print(" - catalog:", catalog_path)
print(" - intensities:", intensities_path)
print(" - merged max PGA:", merged_path)
```

```python
# Optional Cell 10 - empirical exceedance frequency for max PGA across exposures
# For a single-year catalog this is per-catalog; consider Monte Carlo over many catalogs later.

vals = merged['max_pga_g'].dropna().sort_values(ascending=False).reset_index(drop=True)
n = len(vals)
# plotting PGA thresholds and fraction exceedance
pga_thresholds = np.logspace(np.log10(max(vals.min(), 1e-6)), np.log10(vals.max()+1e-6), num=50)
exceedance_frac = [(vals > thr).sum() / n for thr in pga_thresholds]

plt.figure(figsize=(7,5))
plt.loglog(pga_thresholds, exceedance_frac, marker='o')
plt.xlabel('PGA threshold (g)')
plt.ylabel('Fraction of exposures exceeding (empirical)')
plt.title('Empirical exceedance fraction â€” exposures')
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.show()
```
