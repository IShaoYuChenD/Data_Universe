```python
# Notebook cell: Imports, env check, seed, quick demo of helpers

import sys, os
# ensure project root is on sys.path (we run the notebook from notebooks/ or repo root)
repo_root = os.path.abspath(os.path.join(".."))
if repo_root not in sys.path:
    sys.path.insert(0, repo_root)

import platform
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Python:", platform.python_version())
print("numpy:", np.__version__)
print("pandas:", pd.__version__)

# Import helper functions from src
from src.exposure import sample_lat_lon_within_bbox, sample_replacement_values

# reproducibility seed
SEED = 42

# Quick parameter example: small bbox (example: approx. bounding box for a part of California)
example_bbox = (34.0, -118.7, 34.3, -118.2)  # (min_lat, min_lon, max_lat, max_lon)

# Demo: sample 5 coordinates and 5 replacement values
coords_df = sample_lat_lon_within_bbox(n=5, bbox=example_bbox, seed=SEED)
values_series = sample_replacement_values(n=5, distribution_params=None, seed=SEED)

print("\nSample coordinates:")
print(coords_df)
print("\nSample replacement values (currency units):")
print(values_series)

# Small scatter plot demo
plt.figure(figsize=(5,4))
plt.scatter(coords_df['longitude'], coords_df['latitude'], s=30)
plt.title("Sample property coordinates (demo)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```

```python
# Notebook cell: Generate full synthetic exposure, quick EDA, save CSV + metadata

import os
from IPython.display import display
import matplotlib.pyplot as plt

# Import generator (already implemented in src/exposure.py)
from src.exposure import generate_synthetic_exposure, save_metadata

# ----- PARAMETERS (edit if you want different sample size / region) -----
N = 1000
REGION_BBOX = (34.0, -121.5, 38.5, -118.7)  # corrected bbox (min_lat, min_lon, max_lat, max_lon)
BUILDING_TYPE_PROBS = {"wood": 0.75, "masonry": 0.20, "rc": 0.05}
VALUE_DISTR = {"type": "lognormal", "mu": 12.0, "sigma": 0.8}
YEAR_BUILT_RANGE = (1900, 2020)
DEDUCTIBLE_PCT = 0.01
SEED = 42
OUTPUT_DIR = "../data"
CSV_PATH = os.path.join(OUTPUT_DIR, "exposure_sample.csv")
META_PATH = os.path.join(OUTPUT_DIR, "exposure_sample_meta.json")
# -------------------------------------------------------------------------

# Generate dataset
df = generate_synthetic_exposure(
    n=N,
    region_bbox=REGION_BBOX,
    building_type_probs=BUILDING_TYPE_PROBS,
    value_distribution=VALUE_DISTR,
    year_built_range=YEAR_BUILT_RANGE,
    deductible_pct=DEDUCTIBLE_PCT,
    occupancy="residential",
    seed=SEED
)

# Quick checks / EDA
print("Rows:", df.shape[0])
display(df.head())

print("\nColumn types and null counts:")
print(df.info())

print("\nBuilding type counts:")
print(df['building_type'].value_counts())

print("\nReplacement value summary:")
print(df['replacement_value'].describe().round(2))

# Histogram of replacement values (log scale for readability)
plt.figure(figsize=(6,4))
plt.hist(df['replacement_value'], bins=50)
plt.title("Distribution of replacement values (sample)")
plt.xlabel("Replacement value (currency units)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Scatter map of coordinates
plt.figure(figsize=(6,5))
plt.scatter(df['longitude'], df['latitude'], s=6, alpha=0.6)
plt.title("Property coordinates (synthetic sample)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()

# Save CSV and metadata
os.makedirs(OUTPUT_DIR, exist_ok=True)
df.to_csv(CSV_PATH, index=False)

metadata = {
    "n": N,
    "region_bbox": REGION_BBOX,
    "building_type_probs": BUILDING_TYPE_PROBS,
    "value_distribution": VALUE_DISTR,
    "year_built_range": YEAR_BUILT_RANGE,
    "deductible_pct": DEDUCTIBLE_PCT,
    "seed": SEED,
    "notes": "Synthetic dataset for CatRisk project. Units: currency units (user-defined)."
}
save_metadata(metadata, META_PATH)

print(f"\nSaved sample CSV to: {CSV_PATH} ({os.path.getsize(CSV_PATH)} bytes)")
print(f"Saved metadata JSON to: {META_PATH} ({os.path.getsize(META_PATH)} bytes)")

```
